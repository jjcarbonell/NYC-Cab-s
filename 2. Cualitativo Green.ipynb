 **7. Calculamos las relación tiempo/km más alta y más baja**

```python
dg3.dtypes
```

Necesitamos restar el tiempo de bajada de bandera con recogida y pasarlo a segundos y mas tarde a minutos, para los calculos que necesitamos realizar

```python
dg3['Time_carreer_seconds'] = dg3.apply(
    lambda s:
    time.mktime((datetime.strptime(s['Lpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S")).timetuple())
    -time.mktime((datetime.strptime(s['lpep_pickup_datetime'],"%Y-%m-%d %H:%M:%S")).timetuple())
    ,axis=1)
```

```python
dg3.head()
```

```python
dg3['Time_carrer_minutes'] = dg3.apply(
    lambda s:
    (s['Time_carreer_seconds'] % 3600) // 60
    ,axis=1)
```

```python
dg3.head()
```

Ahora podemos realizar el cálculo de minutos respecto a los km o millas recorridos

```python
dg3['Time_long_vs_km'] = dg3.apply(
    lambda s: round(s['Time_carreer_seconds']/s['Trip_distance'],2), axis=1)
```

```python
expensive = dg3[dg3['Time_long_vs_km'] ==  np.max(dg3['Time_long_vs_km'])]
```

 **8. ¿Cuál es el trayecto en el que la relación tiempo/km es más alta?**

```python
expensive
```

```python
print(expensive['Trip_distance'])
print(expensive['Time_carrer_minutes'])
```

## 2.3 Muestra la evolución del tiempo medio de trayecto a lo largo del día. Muestra la evolución de la distancia media de trayecto a lo largo del día.

**1. Creamos variables relevantes para el caso de estudio**

```python
dg3['Week_day'] = dg3.apply(
    lambda s:
    datetime.weekday(datetime.strptime(s['Lpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S"))+1
    ,axis=1)
```

```python
dg3.head()
```

```python
dg3['Hour_day'] = dg3.apply(
    lambda s:
    datetime.strptime(s['Lpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S").hour
    ,axis=1)
```

```python
dg3.head()
```

```python
def df(d,dia,hora):

    return d[(d['Week_day']==dia) & (d['Hour_day']==hora)]
```

### Realizamos una estimacion de la media de tiempo y recorrido por dia de la semana y hora.

```python
dia=1
hora=0

while dia<8:
    while hora<24:
        data=(df(dg3,dia,hora))
        print('Dia: ',dia, 'Hora: ',hora,'Tiempo: ',round(sum(data['Time_carrer_minutes'])/len(data),2),'Recorrido: ',round(sum(data['Trip_distance'])/len(data),2))
        hora=hora+1
    dia=dia+1
    hora=0

#sns.catplot(x="Week_day", y="Hour_day", col=avg, data=data)    
```
