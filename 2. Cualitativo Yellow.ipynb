 **7. Calculamos las relación tiempo/km más alta y más baja**

```python
dy3.dtypes
```

```python
dy3['Time_carreer_seconds'] = dy3.apply(
    lambda s:
    time.mktime((datetime.strptime(s['tpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S")).timetuple())
    -time.mktime((datetime.strptime(s['tpep_pickup_datetime'],"%Y-%m-%d %H:%M:%S")).timetuple())
    ,axis=1)
```

```python
dy3.head()
```

```python
dy3['Time_carrer_minutes'] = dy3.apply(
    lambda s:
    (s['Time_carreer_seconds'] % 3600) // 60
    ,axis=1)
```

```python
dy3.head()
```

```python
dy3['Time_long_vs_km'] = dy3.apply(
    lambda s: round(s['Time_carreer_seconds']/s['trip_distance'],2), axis=1)
```

```python
expensive = dy3[dy3['Time_long_vs_km'] ==  np.max(dy3['Time_long_vs_km'])]
```

 **8. ¿Cuál es el trayecto en el que la relación tiempo/km es más alta?**

```python
print(expensive['trip_distance'])
print(expensive['Time_carrer_minutes'])
```

## 2.3 Muestra la evolución del tiempo medio de trayecto a lo largo del día. Muestra la evolución de la distancia media de trayecto a lo largo del día.

**1. Creamos variables relevantes para el caso de estudio**

```python
dy3['Week_day'] = dy3.apply(
    lambda s:
    datetime.weekday(datetime.strptime(s['tpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S"))+1
    ,axis=1)
```

```python
dy3.head()
```

```python
dy3['Hour_day'] = dy3.apply(
    lambda s:
    datetime.strptime(s['tpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S").hour
    ,axis=1)
```

```python
dy3.head()
```

```python
def df(d,dia,hora):

    return d[(d['Week_day']==dia) & (d['Hour_day']==hora)]
```

```python
dia=1
hora=0

while dia<8:
    while hora<24:
        data=(df(dy3,dia,hora))
        print('Dia: ',dia, 'Hora: ',hora,'Tiempo: ',round(sum(data['Time_carrer_minutes'])/len(data),2),'Recorrido: ',round(sum(data['trip_distance'])/len(data),2))
        hora=hora+1
    dia=dia+1
    hora=0

 
```

Observamos que los tiempos son altos para el poco recorrido que se hace, en comparación con los datos de taxis Green
